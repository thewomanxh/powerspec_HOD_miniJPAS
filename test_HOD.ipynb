```python
import numpy as np
import powerspec_HOD as hod
import matplotlib.pyplot as plt
```

```python
ell_interp = np.logspace(np.log10(200), np.log10(20000), 10)
zz_inte = np.linspace(0.2, 0.6, num = 20, endpoint = True)
ibin_num=1
M_min_fit = np.linspace(10, 13, 10)
M_1_fit = np.linspace(11, 15, 20)
alpha_fit = np.linspace(0.5, 2, 10)
```

```python
fig=plt.figure(1,(8,5))
plt.loglog()
for i in range(len(M_min_fit)):
    th_interp = hod.TheoryDATA(10.**M_min_fit[i], 10.**M_1_fit[0], alpha_fit[0], zz_inte, ell_interp, ibin_num)
    plt.plot(ell_interp, th_interp, color=(i/len(M_1_fit), 0.2 ,0.5))
plt.ylim(1.e-9, 1.e-4)
plt.xlim(200, 20000)
plt.tight_layout()
plt.show
```

```python
fig=plt.figure(1,(8,5))
plt.loglog()
for i in range(len(M_1_fit)):
    th_interp = hod.TheoryDATA(10.**M_min_fit[0], 10.**M_1_fit[i], alpha_fit[0], zz_inte, ell_interp, ibin_num)
    plt.plot(ell_interp, th_interp, color=(i/len(M_1_fit), 0.2 ,0.5))
plt.ylim(1.e-10, 1.e-4)
plt.xlim(200, 20000)
plt.tight_layout()
plt.show
```

```python
fig=plt.figure(1,(8,5))
plt.loglog()
for i in range(len(alpha_fit)):
    th_interp = hod.TheoryDATA(10.**M_min_fit[0], 10.**M_1_fit[0], alpha_fit[i], zz_inte, ell_interp, ibin_num)
    plt.plot(ell_interp, th_interp, color=(i/len(M_1_fit), 0.2 ,0.5))
plt.ylim(1.e-10, 1.e-3)
plt.xlim(200, 20000)
plt.tight_layout()
plt.show
```
